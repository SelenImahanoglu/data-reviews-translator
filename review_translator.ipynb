## Review Translator

ğŸ‘‰ `reviews` datasetâ€™ini load edin ve 1-yÄ±ldÄ±z ratingâ€™e sahip reviewâ€™lardan bir Ã¶rnek (sample) seÃ§in.

```python
# Verileri yÃ¼kle
from olist.data import Olist
data = Olist().get_data()
```

ğŸ‘€ 20 adet dÃ¼ÅŸÃ¼k puan alan yorumdan oluÅŸan bir Ã¶rneklem seÃ§in (rastgele) ve bunu bir listeye dÃ¶nÃ¼ÅŸtÃ¼rÃ¼n.

```python
# 1. TÃ¼m yorumlarÄ± veriden alalÄ±m
reviews = data['order_reviews']

# 2. Sadece 1 yÄ±ldÄ±zlÄ± olanlarÄ± ve yorum kÄ±smÄ± boÅŸ (NaN) OLMAYANLARI filtreleyelim
bad_reviews = reviews[(reviews['review_score'] == 1) & (reviews['review_comment_message'].notnull())]

# 3. Rastgele 20 tanesini seÃ§ip (sample) sadece metinleri bir listeye (tolist) Ã§evirelim
sample_reviews = bad_reviews['review_comment_message'].sample(n=20, random_state=42).tolist()

# SeÃ§ilen ilk 3 yoruma gÃ¶z atalÄ±m (Portekizce)
sample_reviews[:3]
```

ğŸ—£ Bu gÃ¶reve baÅŸlamadan Ã¶nce Ã¶nceden yÃ¼klediÄŸiniz `google_translator` paketini kullanarak bu yorumlarÄ± Ã§evirin:

```python
from deep_translator import GoogleTranslator

# Portekizce'den (pt) Ä°ngilizce'ye (en) Ã§evirecek objemizi oluÅŸturuyoruz
translator = GoogleTranslator(source='pt', target='en')
translated_reviews = []

print("Ã‡eviri iÅŸlemi baÅŸladÄ±, lÃ¼tfen bekleyin...")

for text in sample_reviews:
    try:
        # Yeni Ã§evirmenimizle metni Ã§eviriyoruz
        translation = translator.translate(text)
        translated_reviews.append(translation)
    except Exception as e:
        translated_reviews.append(f"HATA: {e}")

print("Ã‡eviri tamamlandÄ±!\n")

# Ã‡evrilen ilk 3 yoruma gÃ¶z atalÄ±m
for i in range(3):
    print(f"Orijinal: {sample_reviews[i]}")
    print(f"Ã‡eviri:   {translated_reviews[i]}\n")
```

```python
# TÃ¼m 20 Ã§eviriyi ekrana yazdÄ±rÄ±p dedektifliÄŸe devam edelim
for i, review in enumerate(translated_reviews, 1):
    print(f"[{i}] {review}\n")
```
